```python
import pandas as pd
```

```python
df_train = pd.read_csv("../data/train.csv")
df_test = pd.read_csv("../data/test.csv")
```

```python
rename_column_lambda = lambda columns: [column.strip().lower().replace("-", "_") for column in columns]

df_train.columns = rename_column_lambda(df_train.columns)
df_test.columns = rename_column_lambda(df_test.columns)
```
